# Reading data from the HTML file
Extract a tr tag (table row) from the html file 

```python
row = soup.find("tr")
print(row)
```

That gave us the first row which contains table headers, but we want all rows 

```python
rows = soup.findAll("tr")
print(rows)
```

Now with findall we get a list of row's, I was lazy in my previous lab's table definition and only gave it two rows

```python
#loop through each row, get td data and print it
#this will ignore th tag
for row in rows:
    cols = row.findAll("td")
    for col in cols:
        print(col.text)
```

# Tie the bits together

```python
#open the url using beautifulSoup html.parsr
with open("../Week 2 - Javascript/lab2.html") as fp: 
    soup = BeautifulSoup(fp,'html.parser') 
    
#Open csv file to write to it (creates new if not exists)
with open('week02data.csv', mode='w', newline='') as employee_file:
    employee_writer = csv.writer(employee_file, delimiter=',', quotechar='"', quoting = csv.QUOTE_MINIMAL)
    #findall rows with tr tag
    rows = soup.findAll("tr")
    for row in rows:
        dataList=[] #list object to hold all text in td tags
        cols = row.findAll("td")
        for col in cols:
            #if col has button in it don't add to list
            if not col.find("button"):
                dataList.append(col.text)
        #write to csv if list not empty
        if dataList:
            #write list to csv once have looped through every column
            employee_writer.writerow(dataList)
        #continue loop across all rows
```

```python
page = requests.get("https://www.myhome.ie/residential/mayo/property-for-sale?page=1") 
soup = BeautifulSoup(page.content, 'html.parser') 

listings = soup.findAll("div", class_="PropertyListingCard")

with open("week03_home.csv", mode='w', newline='') as home_file:
    home_writer = csv.writer(home_file,delimiter='\t', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    
    for listing in listings:
        entryList = []
        price=listing.find(class_="PropertyListingCard__Price").text
        entryList.append(price.strip())
        address = listing.find(class_="PropertyListingCard__Address").text
        entryList.append(address.strip())
        
        home_writer.writerow(entryList)
```
